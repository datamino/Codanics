```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=sns.load_dataset('titanic')
```

```python
df.columns
```

```python
sns.scatterplot(data=df,x='age',y='fare')
plt.show()
```

```python
sns.scatterplot(data=df,x='age',y='fare' ,color='orange')
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.scatterplot(data=df,x='age',y='fare' ,hue='sex',size='fare')
plt.show()
```

```python
# plt.figure(figsize=(10,6))
# sns.scatterplot(data=df,x='age',y='fare',hue='sex',palette=['red','blue'])
# plt.show()
# otherways
sns.set_theme(style='darkgrid',font_scale=1.5)
palette={'male':'orange','female':'black'}
plt.figure(figsize=(10,6))
sns.scatterplot(data=df,x='age',y='fare',hue='sex',palette=palette)
plt.savefig('delete.png',dpi=600)
plt.show()
```

```python
plt.figure(figsize=(16,9))
sns.set_theme(style='darkgrid',font_scale=1.5)
sns.boxplot(data=df,x='sex',y='age',hue='class',palette=['Red','Blue','Orange'])
plt.xlabel('Gender')
plt.ylabel('Age')
plt.title('Age vs Sex Box Plot')
plt.savefig('helllo.png',dpi=600)
plt.legend(loc='upper right',title='Classes')
plt.show()
```

```python
plt.figure(figsize=(16,9))
sns.set_theme(style='darkgrid',font_scale=1.5)
sns.boxenplot(data=df,x='sex',y='age',hue='class',palette=['Red','Blue','Orange'])
plt.xlabel('Gender')
plt.ylabel('Age')
plt.title('Age vs Sex Box Plot')
plt.savefig('helllo.png',dpi=600)
plt.legend(loc='upper right',title='Classes')
plt.xticks(ticks=[0,1],labels=['Male','Female'],rotation=45)
plt.show()
```

```python
plt.figure(figsize=(16,9))
sns.set_theme(style='darkgrid',font_scale=1.5)
sns.barplot(data=df,x='sex',y='age',hue='class',palette=['Red','Blue','Orange'],capsize=0.1)
plt.xlabel('Gender')
plt.ylabel('Age')
plt.title('Age vs Sex Box Plot')
plt.savefig('helllo.png',dpi=600)
plt.legend(loc='upper right',title='Classes')
plt.xticks(ticks=[0,1],labels=['Male','Female'],rotation=45)
plt.show()
```
