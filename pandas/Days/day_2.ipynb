# **About Data Set**


**Description**

This dataset contains demographic information from the Pakistan Population Census conducted in 2017. It provides detailed population data at various administrative levels within Pakistan, including provinces, divisions, districts, and sub-divisions. The dataset also includes information on urban and rural populations, gender distribution, transgender individuals, sex ratios, population figures from the 1998 census, and annual growth rates.

**Features**

`Province:` The administrative provinces or regions of Pakistan where the census data was collected.

`Division:` The divisions within each province. Divisions are the second level of administrative divisions in Pakistan.

`District:` Districts within each division, representing larger administrative units.

`Sub-Division:` Sub-divisions or tehsils within each district, providing more localized data.

`Area: `The land area (in square kilometers) of each sub-division.

`Urban Population 2017:` The population of urban areas within each sub-division for the year 2017.

`Rural Population 2017:` The population of rural areas within each sub-division for the year 2017.

`Male Population 2017:` The male population within each sub-division for the year 2017.

`Female Population 2017:` The female population within each sub-division for the year 2017.

`Transgender Population 2017:` The population of transgender individuals within each sub-division for the year 2017.

`Sex Ratio 2017:` The sex ratio, calculated as the number of females per 1000 males, within each sub-division for the year 2017.

`Population in 1998:` The total population of each sub-division as recorded in the 1998 census.

`Annual Growth Rate:` The annual growth rate of the population in each sub-division, calculated as the percentage increase from 1998 to 2017.

```python
urban_pupulation_in_1998=df['POPULATION 1998 (URBAN)']
rural_pupulation_in_1998=df['POPULATION 1998 (RURAL)']
urban_pupulation_in_2017=df['ALL SEXES (URBAN)']
rural_pupulation_in_2017=df['ALL SEXES (RURAL)']
total_pupulation_in_1998=rural_pupulation_in_1998.sum()+ urban_pupulation_in_1998.sum()
total_pupulation_in_2017=rural_pupulation_in_2017.sum()+ urban_pupulation_in_2017.sum()
population_increase = total_pupulation_in_2017 - total_pupulation_in_1998
percentage_increase = (population_increase / total_pupulation_in_1998) * 100

print('-'*50)
print("\t\tPopulation in 1998")
print('-'*50)
print('\n')
print('Total Urban Population is 1998 was:',urban_pupulation_in_1998.sum())
print('Total Rural Population is 1998 was:',rural_pupulation_in_1998.sum())
print('Total Population in 1998 was:',total_pupulation_in_1998)
print('\n')
print('-'*50)
print("\t\tPopulation in 2017")
print('-'*50)
print('\n')
print('Total Urban Population is 2017 was:',urban_pupulation_in_2017.sum())
print('Total Rural Population is 2017 was:',rural_pupulation_in_2017.sum())
print('Total Population in 2017 was:',total_pupulation_in_2017)
print('-'*50)

print('Total Increase  in Population from 1998 to  2017  was:',population_increase)
print('Total Percentage of  Increase  in Population from 1998 to  2017  was:',percentage_increase)
print('-'*50)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker

# Prepare data for visualization
data = {
    'Location': ['Urban', 'Rural', 'Urban', 'Rural'],
    'Year': ['1998', '1998', '2017', '2017'],
    'Population': [urban_pupulation_in_1998.sum(), rural_pupulation_in_1998.sum(), urban_pupulation_in_2017.sum(), rural_pupulation_in_2017.sum()]
}

# Convert data into DataFrame
df_plot = pd.DataFrame(data)

# Create bar plot
plt.figure(figsize=(10, 6))
bar_plot = sns.barplot(x='Year', y='Population', hue='Location', data=df_plot)
plt.title('Population changes from 1998 to 2017')

# Format y-axis labels to show actual population
bar_plot.yaxis.set_major_formatter(ticker.EngFormatter())

plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
import pandas as pd

# Data
years = ['1998', '2017']
total_population = [total_pupulation_in_1998, total_pupulation_in_2017]

# Create a DataFrame
data = pd.DataFrame({
    'Year': years,
    'Total Population': total_population,
})

# Colors
colors = ['darkorange', 'blue']  # Make orange more colorful

# Bar plot
sns.barplot(x='Year', y='Total Population', data=data, palette=colors, hue='Year', legend=False)

# Create Legend
orange_patch = mpatches.Patch(color='darkorange', label='1998')
blue_patch = mpatches.Patch(color='blue', label='2017')
plt.legend(handles=[orange_patch, blue_patch])

# Set plot title and labels
plt.title('Total Population: 1998 vs 2017')
plt.xlabel('Year')
plt.ylabel('Total Population')

# Show the plot
plt.show()
```
