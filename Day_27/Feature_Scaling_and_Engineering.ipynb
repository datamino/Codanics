3) **Feature/Column Scaling (Normliztion)**
  > Feature Scaling is a method to limit the range of variables so that they can be compared on common grounds.It helps Algorithms to coverage faster and modal perform better ,especially for modals like regression, logistic regression and supprt vector machines.
  
  #### **Methods**
- Min Max Scaling.
- Standard Scaling (Z-Score Normalization)
- Robost Scaling.
- Logarthimic Scaling.


#### **Min Max Scaling.**

X`=X-min(X) / max(X)-min(X)    &nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;       (0<=X<=1)


#### **Standard Scaling.**

X`=X-mean(X) / std(X)   &nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;           (-3<=X<=3)
- Normal distributions follow the empirical rule, also called the 68-95-99.7 rule
    - (-1<=X<=1)   68% Data lie in this range.
    - (-2<=X<=2)   95% Data lie in this range.
    - (-3<=X<=3)   99.7% Data lie in this range.

#### **Robost Scaling.**

X`=X-median(X) / IQR(X)    &nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;       (negtive-<=X<=Postive)

-It gives Values in between  Percetile 25%  to 75% Percentile.

#### **Logarthimic Scaling..**

X`=Take log10 of data.  




  
   

```python
from sklearn.preprocessing import MinMaxScaler
import pandas as pd
import numpy as np
```

```python
data={
    'Value':[10,20,30,40,50]
}
df=pd.DataFrame(data)
df
```

# **Min-Max Scalar**

```python
# Method 1
# scalar=MinMaxScaler()
# df['Scaled Value']=scalar.fit_transform(df['Value'].values.reshape(-1,1))
# df

# Method 2
df['Scaled Value Min Max']=MinMaxScaler().fit_transform(df[['Value']])
df
```

# **Standard Scalar**

```python
from sklearn.preprocessing import StandardScaler

# Method 1
# scalar=StandardScaler()
# df['Scaled Value']=scalar.fit_transform(df['Value'].values.reshape(-1,1))
# df

# Method 2
df['Scaled Value Standard']=StandardScaler().fit_transform(df[['Value']])
df
```

# **Robost Scalar**

```python
from sklearn.preprocessing import RobustScaler
# scalar=RobustScaler()
# df['Scaled Value Robost']=scalar.fit_transform(df['Value'].values.reshape(-1,1))
# df

# Method 2
df['Scaled Value Robost']=RobustScaler().fit_transform(df[['Value']])
df
```

# **Logarthimic Scalar**

```python

df['Scaled Value Logartimic']=np.log(df['Value'])
df['Scaled Value Logartimic2']=np.log(df['Value'])
df['Scaled Value Logartimic10']=np.log(df['Value'])

df
```
